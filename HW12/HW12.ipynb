## K-Means Clustering

## 1. Numerical Data

```python
import pandas as pd
data=pd.read_csv("iris.csv")
data.head()
```

```python
x= data.iloc[:,3:5]
x
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['PetalLengthCm'],data['PetalWidthCm'])
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(3)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['PetalLengthCm'],data_with_clusters['PetalWidthCm'],c=data_with_clusters['Clusters'],cmap='rainbow')
plt.show()
```

## Finding the optimal K using WCSS(Within-Cluster Sum of Square)

KMeans(n_clusters=8, *, init='k-means++', n_init=10, max_iter=300, tol=0.0001, verbose=0, random_state=None, copy_x=True, algorithm='auto')

```python
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
import cv2
# read the image
image = cv2.imread("image.jpg")
```

```python
# convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)
```

```python
import numpy as np
# reshape the image to a 2D array of pixels and 3 color values (RGB)
pixel_values = image.reshape((-1, 3))
# convert to float
pixel_values = np.float32(pixel_values)
```

```python
print(pixel_values.shape)
```

```python
# define stopping criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
```

```python
# number of clusters (K)
k = 3
_, labels, (centers) = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
# convert back to 8 bit values
centers = np.uint8(centers)

# flatten the labels array
labels = labels.flatten()
```

```python
# convert all pixels to the color of the centroids
segmented_image = centers[labels.flatten()]
```

```python
import matplotlib.pyplot as plt
# reshape back to the original image dimension
segmented_image = segmented_image.reshape(image.shape)
# show the image
plt.imshow(segmented_image)
plt.show()
```

```python
# disable only the cluster number 2 (turn the pixel into black)
masked_image = np.copy(image)
# convert to the shape of a vector of pixel values
masked_image = masked_image.reshape((-1, 3))
# color (i.e cluster) to disable
cluster = 2
masked_image[labels == cluster] = [0, 0, 0]
# convert back to original shape
masked_image = masked_image.reshape(image.shape)
# show the image
plt.imshow(masked_image)
plt.show()
```
